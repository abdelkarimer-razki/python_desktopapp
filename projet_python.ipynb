```python
def run():
    address_book = AddressBook()
    while True:
        print("1. Add Contact")
        print("2. Remove Contact")
        print("3. Display Contacts")
        print("4. Quit")
        choice = input("Enter your choice: ")
        if choice== "1":
            name = input("Enter name: ")
            email = input("Enter email: ")
            phone_number = input("Enter phone number: ")
            address_book.add(name, email, phone_number)
        elif choice == "2":
            name = input("Enter name: ")
            address_book.delete(name)
        elif choice == "3":
            address_book.show()
        elif choice == "4":
            break 


class contact:
    def __init__(self ,nom, email, phone):
        self.nom = nom
        self.email = email
        self.phone = phone

class AddressBook:
    def __init__(self):
        self.list_contacts = []

    def add(self, nom, email, phone):
        ct = contact(nom, email, phone)
        self.list_contacts.append(ct)
    
    def delete(self, name):
        self.list_contacts = [contact for contact in self.list_contacts if contact.nom != name]
    
    def show(self):
        for contact in self.list_contacts:
            print("======================")
            print("nom:", contact.nom)
            print("email:", contact.email)
            print("phone:", contact.phone)
    

run()
```

```python
import json


class contact:
    def __init__(self ,nom, email, phone):
        self.nom = nom
        self.email = email
        self.phone = phone

class AddressBook:
    def __init__(self):
        self.list= []
        self.filename = "AddressBook.json"

    def add(self, nom, email, phone):
        try:
            with open(self.filename, "r") as file:
                self.list = json.load(file)
        except:
            self.list = []
        self.list.append(contact(nom, email, phone).__dict__)
        with open(self.filename, "w") as file:
            json.dump(self.list, file, indent=4)
    
    def delete(self, nom):
        try:
            with open(self.filename, "r") as file:
                data = json.load(file)
            
            data = [contact for contact in data if contact["nom"] != nom]
            with open(self.filename, "w") as file:
                json.dump(data, file, indent=4)
        except:
            print("No contact is yet stored")


    def show(self):
        try:
            with open(self.filename, "r") as file:
                data = json.load(file)
        except:
            print("No contact is yet stored")
        print(data)


run()

```
